## Operadores Lógicos Bit-a-bit (`|` e `&`)

Este código ilustra o uso dos operadores lógicos bit-a-bit `|` (OR/OU) e `&` (AND/E). Embora funcionem com booleanos, eles operam a nível de bits e podem ter um comportamento diferente dos operadores `or` e `and` quando usados com números ou avaliações de curto-circuito.

```python
# inativo é definido como False.
# A variável ativo (pelo contexto do notebook) é True.
# Quando usamos operadores bit-a-bit (|, &) com valores booleanos, o Python os trata como números inteiros: True como 1 e False como 0.
# Então, a condição (ativo | inativo) é avaliada como (True | False), o que para o operador bit-a-bit se torna (1 | 0).
# A operação bit-a-bit 1 | 0 (OR bit-a-bit) resulta em 1.
# Como 1 é avaliado como True em um contexto booleano, a condição do if é verdadeira.
# Portanto, o código imprime "Ativo".

inativo = False
if (ativo | inativo): # ativo = True (1) OU inativo = False (0) = TRUE = ATIVO
    print("Ativo")
else:
    print("Inativo")

# inativo é novamente definido como False.
# A condição (ativo & inativo) é avaliada como (True & False), o que para o operador bit-a-bit se torna (1 & 0).
# A operação bit-a-bit 1 & 0 (AND bit-a-bit) resulta em 0.
# Como 0 é avaliado como False em um contexto booleano, a condição do if é falsa.
# Portanto, o código imprime "Inativo".

inativo = False
if (ativo & inativo): # ativo = True (1) E inativo = False (0) = FALSE = INATIVO
    print("Ativo")
else:
    print("Inativo")
```

## Operadores Lógicos (`or`, `and`) vs. Bit-a-bit (`|`, `&`)

### Operadores Lógicos (`or`, `and`)

*   **Propósito**: Estes operadores são usados para combinar expressões booleanas (que resultam em `True` ou `False`). Eles avaliam a **verdade** das expressões.
*   **Comportamento de Curto-Circuito**: Esta é a característica mais importante. Eles avaliam as expressões da esquerda para a direita e param assim que o resultado pode ser determinado.
    *   `or`: Se a primeira expressão for `True`, ele nem sequer avalia a segunda expressão, pois o resultado já será `True`. Se a primeira for `False`, ele avalia a segunda.
    *   `and`: Se a primeira expressão for `False`, ele nem sequer avalia a segunda expressão, pois o resultado já será `False`. Se a primeira for `True`, ele avalia a segunda.
*   **Retorno**: Não necessariamente retornam `True` ou `False`. Eles retornam o valor da última expressão avaliada que determinou o resultado.
    *   Exemplo: `False or 'hello'` retorna `'hello'` (porque `False` não foi suficiente, então avalia `hello` e retorna `hello`).
    *   Exemplo: `True and 10` retorna `10`.

### Operadores Bit-a-bit (`|`, `&`)

*   **Propósito**: Estes operadores trabalham diretamente com a **representação binária (bits)** dos números inteiros (ou booleanos, que são tratados como `0` e `1`). Eles realizam operações bit a bit.
*   **Sem Curto-Circuito**: Eles sempre avaliam **ambos os lados** da expressão, independentemente do resultado parcial.
*   **Retorno**: Retornam um novo número inteiro (ou `True`/`False` se os operandos forem booleanos) que é o resultado da operação bit-a-bit.
    *   `|` (OR Bit-a-bit): Compara cada bit correspondente. Se pelo menos um dos bits for `1`, o bit resultante é `1`. Caso contrário, é `0`.
    *   `&` (AND Bit-a-bit): Compara cada bit correspondente. Se ambos os bits forem `1`, o bit resultante é `1`. Caso contrário, é `0`.

```python
a = 10
b = 20

if (a > 11) or (b > 19): # boa pratica nas condicionais complexas tu separar cada uma dentro de um parenteses "()"
    print("or VERDADEIRO")
else:
    print("or FALSO")

# Primeiro, o operador bit-a-bit & é aplicado aos números 11 e b (que é 20).

if a > 11 | b > 19:
    print("| VERDADEIRO")
else:
    print("| FALSO")

if a < 11 and b > 19:
    print("and VERDADEIRO")
else:
    print("and FALSO")

if (a < 11) & (b > 19):
    print("& VERDADEIRO")
else:
    print("& FALSO")

# Primeiro, o operador bit-a-bit & é aplicado aos números 11 e b (que é 20).

# 11 em binário é 01011
# 20 em binário é 10100
# 11 & 20 (bit-a-bit AND) resulta em 00000, que é 0 em decimal.
# Depois, essa expressão se torna uma comparação encadeada: a < 0 > 19.

# O Python avalia comparações encadeadas da esquerda para a direita, como se fosse (a < 0) and (0 > 19).
# a < 0 se torna 10 < 0, que é False.
# 0 > 19 é False.
# Como False e False é False, a condição completa a < 0 > 19 avalia como False.


if b > 19 | a < 9:
    print("& sim")
else:
    print("& não ")
```

O loop `while` executa repetidamente um bloco de código enquanto uma condição especificada for verdadeira. A execução do loop continua até que a condição se torne falsa.

```python
contagem = 0
while contagem <= 5:
    print(f"Número: {contagem}")
    # contagem = contagem + 1
    contagem += 1
```

```python
print("\n--- Exemplo 2: Contagem Regressiva ---")

contador_regressivo = 5
while contador_regressivo > 0:
    print(f"Faltam {contador_regressivo} segundos...")
    contador_regressivo -= 1
print("Lançamento!")
```

## BREAK E CONTINUE

### Explicação: `break` e `continue`

- O comando `break` é usado para **interromper completamente** a execução do loop atual, saindo dele.
- O comando `continue` é usado para **pular a iteração atual** do loop e prosseguir para a próxima iteração.

```python
for temp in [20, 25, 999, 30]: # 999 é um erro do sensor!
    if temp == 999:
        print("Dado corrompido, pulando...")
        continue # Ignora o 999 e vai para o 30

    print(f"Processando temperatura: {temp}")
```

```python
condicao = 0
while True:
  print("Contagem", condicao)
  if condicao == 10:
    break;
  condicao += 1
```

```python
for i in range(10):
    if i == 3:
        print('pulando...')
        continue  # Pula o 3 e segue
    if i == 7:
        break     # Para tudo ao chegar no 7
    print(i)
```

## LIST COMPREHENSION

### Explicação: List Comprehension

List Comprehension é uma maneira concisa e eficiente de criar listas em Python. Ela permite gerar uma nova lista aplicando uma expressão a cada item de uma sequência existente, muitas vezes em uma única linha de código.

```python
# list comprehension iterando listas

# precos = [100,200,300]

# preco_com_taxa = []
# for p in precos:
#   preco_com_taxa.append(p * 1.1)
# preco_com_taxa

# list comprehension
precos = [100, 200, 300]

preco_com_taxa = [p * 1.1 for p in precos]
print(preco_com_taxa)
```

```python
# list comprehension iterando tuplas
temperaturas_celsius = (0,10,20,30,40)

temperatura_fahrenheit = [((9/5)*temp + 32) for temp in temperaturas_celsius]
temperatura_fahrenheit
```

```python
# list comprehension iterando em um range manual
quadrados = [x**2 for x in range(1,11)] # vai criar uma lista de resultados do x ao quadrado (x**2) de cada numero entre (range) 1 e 10
quadrados
```
